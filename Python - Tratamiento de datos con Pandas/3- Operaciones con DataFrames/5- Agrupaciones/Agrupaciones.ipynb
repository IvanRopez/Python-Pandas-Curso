```python
### Agregaciones
```

```python
house.groupby('condition').agg({ #diferentes operaciones para cada columna
    'price':'mean',
    'bathrooms': 'median',
    'waterfront': 'max',
    'floors': 'min'
})
```

```python
house.groupby('condition').agg({ #diferentes operaciones para una misma columna
    'price':['mean', 'min', 'max'],
    'bathrooms': 'median',
    'waterfront': 'max',
    'floors':['min','max']
})
```

```python
agrupacion = house.groupby('condition').agg({ #diferentes operaciones para una misma columna
    'price':['mean', 'min', 'max'],
    'bathrooms': 'median',
    'waterfront': 'max',
    'floors':['min','max']
})
agrupacion
```

```python
agrupacion['price']
```

```python
agrupacion['price']['min'].to_frame()
```

```python
agrupacion
```

```python
agrupacion.stack(level = 0)
```

```python
def media_mayores_1_millon(x):
    return np.mean(x[x>1_000_000])
```

```python
house[['price', 'condition']].groupby('condition')['price'].apply(media_mayores_1_millon).to_frame()
```
