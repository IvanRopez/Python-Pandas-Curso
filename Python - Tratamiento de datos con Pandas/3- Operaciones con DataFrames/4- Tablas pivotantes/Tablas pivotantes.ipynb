```python
#Consisten en convertir los valores de las filas en columnas
import numpy as np
import pandas as pd
from pathlib import Path
path = Path('../datos')
pivot = pd.read_csv(path / 'pivot_data.csv')
pivot
```

```python
pivot[pivot['variable'] == 'A']
```

```python
_ = pivot.pivot(index = 'date', columns = 'variable', values = 'value')
_
```

```python
_ = pivot.pivot(index = 'variable2', columns = 'variable', values = 'value')
_
```

```python
#Estableciendo una lista en index se puede hacer multiindex
_ = pivot.pivot(index = ['date','variable'], columns = 'variable2', values = 'value')
_
```

```python
## Melt
```

```python
#Convertir el nombre de variables en el dominio
data_cal = pd.DataFrame({'nombre':['Antonio','Maria'],
                         'apellido':['Lopez','Jimenez'],
                         'nota':[7, 8.9],
                         'num_asignaturas':[3, 5]})
data_cal
```

```python
data_cal.melt(id_vars=['nombre','apellido']) #Nota y num_asig han pasado a ser valores, y ahora hay 2 tablas para cada persona
```
