```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
from pathlib import Path
path = Path('../datos')
house = pd.read_csv(path / 'house.csv')
house.head(2)
```

```python
date = house.drop_duplicates(subset=['date'], keep='first') #Date Ãºnica sin duplicados
date.shape
```

```python
date.head()
```

```python
date.index = date['date'] #Date como Ã­ndice
```

```python
date = date.drop(labels = ['date'], axis=1) #Borrar la columna date porque la tenemos como Ã­ndice
date.head()
```
