```python
'''
Convierte la columna date en un tipo fecha
Genera 3 nuevas columnas fruto de extraer año, mes y día
¿Cuál es el precio medio por mes?
Crea una nueva columna fruto de sumar 20 días a la columna fecha
'''
```

```python
house = pd.read_csv(filepath_or_buffer = path / 'house.csv', sep=',')
house
```

```python
house['date'] = pd.to_datetime(house['date'], yearfirst=True)
house['date']
```

```python
house['year'] = house['date'].dt.year
house['month'] = house['date'].dt.month
house['day'] = house['date'].dt.day
```

```python
house.head(2)
```

```python
house.set_index(house['date']).resample('1ME')['price'].mean()
```

```python
td = pd.Timedelta('20 days')
house['twenty'] = house['date'] + td
house[['date', 'twenty']]
```
